## Retrieving Pokemon Data from the PokeAPI

```python
#Hitting the API and pulling down data for all pokemon in the specified range in json format
#Data here includes information on generation, battle stats, types, physical characteristics...
base_url = "https://pokeapi.co/api/v2/pokemon/"
pokemon = []
for i in range(1,899):
    #Printing a status so we can make sure data is being pulled
    print(i/898*100)
    req = requests.get(base_url + str(i))
    pokemon.append(req.json())
```

```python
#Data for all 898 pokemon
#Making a list
pokemon_all=[]


for entry in pokemon:
    pokemon_dict={}
    pokemon_dict["name"]=entry["name"]
    pokemon_dict["id"]=entry["id"]
    pokemon_dict["height"]=entry["height"]
    pokemon_dict["weight"]=entry["weight"]
    pokemon_dict["base_experience"]=entry["base_experience"]
    pokemon_dict["stats_hp"]=entry["stats"][0]["base_stat"]
    pokemon_dict["stats_attack"]=entry["stats"][1]["base_stat"]
    pokemon_dict["stats_defense"]=entry["stats"][2]["base_stat"]
    pokemon_dict["stats_special_attack"]=entry["stats"][3]["base_stat"]
    pokemon_dict["stats_special_defense"]=entry["stats"][4]["base_stat"]
    pokemon_dict["stats_speed"]=entry["stats"][5]["base_stat"]
    #Pulling out the types
    types=entry["types"]
    type_array=[]
    for t in types:
        type_name=t["type"]["name"]
        type_array.append(type_name)
    pokemon_dict["types"]=type_array
    #Figuring out and including how many types each pokemon represents
    num_types=len(types)
    pokemon_dict["number_types"]=num_types
    #Some pokemon have multiple types, so this pulls out the the first type for all (all have at least 1 type)
    pokemon_dict["type1"]=types[0]["type"]["name"]
    #The max types is 2, so if a pokemon has 2 types, this will put the 2nd type into a new column, otherwise a value of 'none' is given
    if num_types==2:
        pokemon_dict["type2"]=types[1]["type"]["name"]
    else:
        pokemon_dict["type2"]="none"
    #Sorting the array of types by alphabetical order to make manipulation of types easier later
    type_array.sort()
    #Joining the array to create a hybrid "type" column so if there are 2 types they are given together as string
    pokemon_dict["hybrid_type"]=','.join(type_array)
    #Making a new column to identify generation
    if entry["id"] <= 151:
        generation = 1
    elif entry["id"] <= 251:
        generation = 2
    elif entry["id"] <= 386:
        generation = 3
    elif entry["id"] <= 493:
        generation = 4
    elif entry["id"] <= 649:
        generation = 5
    elif entry["id"] <= 721:
        generation = 6
    elif entry["id"] <= 809:
        generation = 7
    else:
        generation = 8
    pokemon_dict["generation"] = generation
    pokemon_all.append(pokemon_dict)

#Making the Data Frame
pokemon_df=pd.DataFrame(pokemon_all)

#Showing the first 5 data frame entries to make sure everything looks as expected.
pokemon_df.head(5)
```

## SQL Queries using Psycopg2

#### Query 1: Looking at growth rate differences amongst Pokemon

```python
#Make connection to the database & set the cursor
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()

#set query request
query ="""
        SELECT s.growth_rate, t.name 
        FROM species AS s 
        JOIN types AS t ON t.name = s.name 
        ORDER BY s.growth_rate;
        """
print(query)
cur.execute(query)
all = cur.fetchall()

print(all)
```

#### Query 2: Looking to see if type is related to habitat (here we look at the waters-edge habitat only)

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()

query = """
        SELECT s.name, t.type1, t.type2, s.habitat
        FROM species AS s
        JOIN types AS t
        ON t.name = s.name
        WHERE s.habitat = 'waters-edge';
        """
print(query)
cur.execute(query)
all = cur.fetchall()
print(all)
```

#### Query 3: Looking at the number of mythical Pokemon in each generation

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()
query= """
        SELECT COUNT(s.is_mythical) AS mythical, s.generation 
        FROM species AS s 
        WHERE s.is_mythical = 'true' 
        GROUP BY generation;
        """

print(query)
cur.execute(query)

all = cur.fetchall()

print(all)
```

#### Query 4: Looking at the number of legendary Pokemon in each generation

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()

query = """
        SELECT COUNT(s.is_legendary) AS legendary, s.generation 
        FROM species AS s 
        WHERE s.is_legendary = 'true' 
        GROUP BY generation;
        """


print(query)
cur.execute(query)

all = cur.fetchall()

print(all)
```

#### Query 5: We hypothesized that poison type Pokemon might be associated with urban habitats and checked to see if that was true

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()

query = """
        SELECT s.name, s.habitat, t.type1, t.type2 
        FROM species AS s 
        JOIN types AS t 
        ON t.name = s.name 
        WHERE s.habitat = 'urban' AND (t.type1 = 'poison' OR t.type2 = 'poison');
        """
print(query)
cur.execute(query)
all = cur.fetchall()
print(all)
```

#### Query 6: Considering if the existence of physical gender differences is related to type

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()
query = """
        SELECT COUNT(s.has_gender_differences) AS gender_diff, COUNT(t.type1), t.type1
        FROM species AS s 
        JOIN types AS t
        ON s.ID = t.ID
        GROUP BY t.type1, has_gender_differences
        ORDER BY gender_diff DESC
    ;
    """
print(query)
cur.execute(query)
all = cur.fetchall()

print(all)

```

#### Query 7: Investigating Type Differences by Generation

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()

query = """
        SELECT COUNT(t.hybrid_type) AS count_type, t.hybrid_type, g.generation
        FROM species AS s
        JOIN types AS t
        ON s.ID=t.ID
        JOIN generation AS g
        ON t.ID=g.ID
        GROUP BY g.generation, t.hybrid_type
        ORDER BY generation, count_type DESC;
        """

print(query)
cur.execute(query)
all = cur.fetchall()

print(all)
```

#### Query 8: Comparing Legendary Base XP to Non-legendary Base XP

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()

query= """
        SELECT (SUM(bs.base_experience)/COUNT(bs.base_experience)) AS average, s.is_legendary
        FROM species AS s
        JOIN battle_stats AS bs
        ON s.ID = bs.ID
        GROUP BY s.is_legendary;
        """
print(query)
cur.execute(query)
all = cur.fetchall()

print(all)
```

#### Query 9:Looking at relationship between switchable forms and type

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()

query = """
        SELECT COUNT(s.forms_switchable) AS forms, t.hybrid_type
        FROM species AS s
        JOIN types AS t
        ON s.ID = t.ID
        GROUP BY t.hybrid_type
        ORDER BY forms DESC;
        """

print(query)
cur.execute(query)
all = cur.fetchall()

print(all)
```

#### Query 10: Considering differences in physical characteristics in different habitats (here, looking at height)

```python
con.close()
con = psycopg2.connect("dbname=pokemon_etl_2 user=postgres")
cur = con.cursor()

query = """
        SELECT (SUM(pc.height)/COUNT(pc.height)) AS avg_height, h.habitat
        FROM physical_characteristics AS pc
        JOIN habitat AS h
        ON pc.ID = h.ID
        GROUP BY h.habitat
        ORDER BY avg_height DESC;
        """


print(query)
cur.execute(query)
all = cur.fetchall()

print(all)
```
