```python
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

engine = create_engine('postgresql://postgres:sp00kysucks@localhost:5432/pokemon_ETL')

Session = sessionmaker(bind=engine)

Base = declarative_base()
```

```python
from sqlalchemy import Column, String, Integer, Date, Boolean

class Types(Base):
    __tablename__ = 'pokemon_types'

    name = Column(String, primary_key=True)
    id = Column(Integer, primary_key=True)
    types = Column(Integer)
    number_types = Column(Integer)
    type1 = Column(String)
    type2 = Column(String)
    hybrid_type = Column(String)
```

```python
from sqlalchemy import Column, String, Integer, Date, Boolean

class Species(Base):
    __tablename__ = 'pokemon_species'

    ID = Column(Integer, primary_key=True)
    name = Column(String, primary_key=True)
    gender_rate = Column(Integer)
    capture_rate = Column(Integer)
    base_happiness = Column(String)
    is_baby = Column(Boolean)
    is_legendary = Column(Boolean)
    is_mythical = Column(Boolean)
    has_gender_differences = Column(Boolean)
    forms_switchable = Column(Boolean)
    growth_rate = Column(String)
    evolves_from_species = Column(String)
    habitat = Column(String)
    generation = Column(String)
```

```python
#Do "urban" Pokemon tend to be poison type?

with engine.connect() as con:
    poison_urban = con.execute('SELECT s.name FROM pokemon_species AS s JOIN pokemon_types AS t ON t.name = s.name WHERE t.type1 = poison OR t.type2 = poison AND s.habitat = urban;')

    for row in poison_urban:
        print(row)
```
